# Definimos funciones para limpiar los df

```python
# Función para reemplazar NaN hasta el primer valor no nulo con "No inscripto"
def fill_na_no_inscripto(row):
    found_first = False
    for col in range(len(row)):
        if pd.notna(row.iloc[col]):
            found_first = True  # Marcamos que encontramos el primer valor no nulo
        elif not found_first:
            row.iloc[col] = "No inscripto"  # Reemplazamos los NaN antes del primer valor no nulo
    return row
```

```python
# Función para reemplazar NaN después de "BAJA" hasta el siguiente valor no nulo
def fill_nan_baja(row):
    found_baja = False
    for col in range(len(row)):
        if found_baja and pd.isna(row.iloc[col]):
            row.iloc[col] = "BAJA"  # Marcamos los NaN después de "BAJA" como "BAJA"
        elif found_baja and pd.notna(row.iloc[col]):
            # Detenemos la marcación después del siguiente valor no nulo
            break
        elif row.iloc[col] == "BAJA":
            found_baja = True  # Marcamos que hemos encontrado "BAJA"
    return row
```

```python
# Función para reemplazar NaN después de "BECA COMPLETA" hasta el siguiente valor no nulo
def fill_nan_beca(row):
    found_beca = False
    for col in range(len(row)):
        if found_beca and pd.isna(row.iloc[col]):
            row.iloc[col] = "BECA"  # Marcamos los NaN después de "BECA COMPLETA" como "BECA COMPLETA"
        elif found_beca and pd.notna(row.iloc[col]):
            # Detenemos la marcación después del siguiente valor no nulo
            break
        elif row.iloc[col] == "BECA":
            found_beca = True  # Marcamos que hemos encontrado "BECA COMPLETA"
    return row
```

```python
def convertir_a_enteros(df):
    for col in df.columns:
        # Iterar sobre cada valor en la columna
        nuevos_valores = []
        for value in df[col]:
            # Si el valor es un número entero, convertirlo a int
            if isinstance(value, str) and value.isdigit():
                nuevos_valores.append(int(value))
            # Si el valor es un número decimal, convertirlo a float
            elif isinstance(value, str) and value.replace('.', '', 1).isdigit():
                nuevos_valores.append(float(value))
            else:
                # Si no es numérico, se deja como está
                nuevos_valores.append(value)
        
        # Reemplazar la columna con los valores convertidos
        df[col] = nuevos_valores
```

```python
def suma_enteros_por_fila(row):
    suma = 0  # Inicializamos la suma en 0
    columnas = ['FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEPT', 'OCT']  # Especificamos las columnas a iterar
    for col in columnas:  # Iteramos sobre las columnas seleccionadas
        value = row[col]  # Obtenemos el valor de la columna actual
        # Verificamos si el valor es un número
        if str(value).isdigit():  # Convertimos a string y comprobamos si es un número
            suma += int(value)  # Convertimos a entero y sumamos
        else:
            # Si no es un número, lo ignoramos
            continue
    return suma  # Devolvemos la suma total de la fila
```

```python
def nombre_columnas(tabla):
    tabla.columns= ['CATEGORIA', 'APELLIDO', 'NOMBRE', 'SEGURO',
       'FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEPT', 'OCT',
       'DEUDA']
```

```python
def mayus(df):
    for col in df.select_dtypes(include="object").columns:
        if col in ["APELLIDO", "NOMBRE"]:
            df[col] = df[col].apply(str.title)
        # Aplicar capitalize a cada valor de la columna si es una cadena de texto
        else:
            df[col] = df[col].apply(str.capitalize)
    return df
```

# Concateno DF

```python
pagos_por_cat_voley = pd.concat([cat_recre, cat_mini, sub12, sub14_f, sub14_m, sub16_f, sub16_m, sub18_f, sub18_m, sub19_f, sub19_m, primera_f, primera_m], axis=0) 
```

```python
pagos_por_cat_voley['CATEGORIA']=pagos_por_cat_voley['CATEGORIA'].replace({
    'Sub-12' :'Sub12',
    'Sub14 fem': 'Sub14', 
    'Sub14 masc': 'Sub14', 
    'Sub16 fem': 'Sub16', 
    'Sub16 masc': 'Sub16', 
    'Sub 18 fem': 'Sub18', 
    'Sub 18 masc': 'Sub18', 
    'Sub 19 fem': 'Sub19', 
    'Sub 19 masc': 'Sub19',
    'Primera fem': 'Primera',
    'Primera masc': 'Primera'
    })
```

```python
pagos_por_cat_voley['CATEGORIA'].unique()
```

```python
pagos_por_cat_voley.to_pickle("pagos_voley_limpio.pkl")
```

```python
# Filtrar las columnas de meses y la categoría
meses = ['FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEPT', 'OCT']
df_meses = pagos_por_cat_voley[['CATEGORIA'] + meses]

# Convertir cada valor en un indicador de beca (1 para 'Beca' y 0 en caso contrario) utilizando .loc para evitar el warning
for mes in meses:
    df_meses.loc[:, mes] = df_meses[mes].apply(lambda x: 1 if x == 'Beca' else 0)


# Agrupar por categoría y sumar los valores de beca para cada mes
becas_por_mes_v = df_meses.groupby('CATEGORIA').sum()

# Transponer el DataFrame para mostrar meses como filas y categorías como columnas
becas_por_mes_v = becas_por_mes_v.T

# Renombrar las filas para que representen mejor cada mes
# becas_por_mes.index.name = 'Mes'
becas_por_mes_v.reset_index(drop=True)
```

```python
# Primero, aseguramos que 'CATEGORIA' sea un índice
becas_por_mes_v = becas_por_mes_v.reset_index()  # Resetea el índice para hacer de los meses una columna regular

# Cambiar el nombre de la columna resultante
becas_por_mes_v.rename(columns={'index': 'MES'}, inplace=True)

# Reorganizar las columnas para que el DataFrame sea más claro
becas_por_mes_v = becas_por_mes_v[['MES'] + [col for col in becas_por_mes_v.columns if col != 'MES']]
```
