```python
from datetime import datetime


def convertir_fecha(fecha):
    # Divide la fecha en día, mes y año
    partes = fecha.split('/')
    
    # Si el mes tiene un solo dígito, añade un "0" al inicio
    if len(partes[1]) == 1:
        partes[1] = '0' + partes[1]
    
    # Si el año tiene dos dígitos, añade "20" al inicio para convertirlo a cuatro dígitos
    if len(partes[2]) == 2:
        partes[2] = '20' + partes[2]
    
    # Une las partes nuevamente en el formato "dd/mm/yyyy"
    fecha_formateada = '/'.join(partes)
    
    # Convierte la fecha al formato "YYYY-MM-DD"
    fecha_convertida = datetime.strptime(fecha_formateada, '%d/%m/%Y').strftime('%Y-%m-%d')
    
    return fecha_convertida
```

```python
# Aplica la función a cada valor de la columna "FECHA"
socios['Fecha_inscripcion'] = socios['Fecha_inscripcion'].apply(convertir_fecha)
socios['Fecha_inscripcion'] = pd.to_datetime(socios['Fecha_inscripcion'])
```

# Definimos funciones para limpiar los df

```python
# Función para reemplazar NaN hasta el primer valor no nulo con "No inscripto"
def fill_na_no_inscripto(row):
    found_first = False
    for col in range(len(row)):
        if pd.notna(row.iloc[col]):
            found_first = True  # Marcamos que encontramos el primer valor no nulo
        elif not found_first:
            row.iloc[col] = "No inscripto"  # Reemplazamos los NaN antes del primer valor no nulo
    return row
```

```python
# Función para reemplazar NaN después de "BAJA" hasta el siguiente valor no nulo
def fill_nan_baja(row):
    found_baja = False
    for col in range(len(row)):
        if found_baja and pd.isna(row.iloc[col]):
            row.iloc[col] = "BAJA"  # Marcamos los NaN después de "BAJA" como "BAJA"
        elif found_baja and pd.notna(row.iloc[col]):
            # Detenemos la marcación después del siguiente valor no nulo
            break
        elif row.iloc[col] == "BAJA":
            found_baja = True  # Marcamos que hemos encontrado "BAJA"
    return row
```

```python
# Función para reemplazar NaN después de "BECA" hasta el siguiente valor no nulo
def fill_nan_beca(row):
    found_beca = False
    for col in range(len(row)):
        if found_beca and pd.isna(row.iloc[col]):
            row.iloc[col] = "BECA"  # Marcamos los NaN después de "BECA" como "BECA"
        elif found_beca and pd.notna(row.iloc[col]):
            # Detenemos la marcación después del siguiente valor no nulo
            break
        elif row.iloc[col] == "BECA":
            found_beca = True  # Marcamos que hemos encontrado "BECA"
    return row
```

```python
def convertir_a_enteros(df):
    for col in df.columns:
        # Iterar sobre cada valor en la columna
        nuevos_valores = []
        for value in df[col]:
            # Si el valor es un número entero, convertirlo a int
            if isinstance(value, str) and value.isdigit():
                nuevos_valores.append(int(value))
            # Si el valor es un número decimal, convertirlo a float
            elif isinstance(value, str) and value.replace('.', '', 1).isdigit():
                nuevos_valores.append(float(value))
            else:
                # Si no es numérico, se deja como está
                nuevos_valores.append(value)
        
        # Reemplazar la columna con los valores convertidos
        df[col] = nuevos_valores
```

```python
def suma_enteros_por_fila(row):
    suma = 0  # Inicializamos la suma en 0
    columnas = ['FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEPT', 'OCT']  # Especificamos las columnas a iterar
    for col in columnas:  # Iteramos sobre las columnas seleccionadas
        value = row[col]  # Obtenemos el valor de la columna actual
        # Verificamos si el valor es un número
        if str(value).isdigit():  # Convertimos a string y comprobamos si es un número
            suma += int(value)  # Convertimos a entero y sumamos
        else:
            # Si no es un número, lo ignoramos
            continue
    return suma  # Devolvemos la suma total de la fila
```

```python
def nombre_columnas(tabla):
    tabla.columns= ['CATEGORIA','APELLIDO','NOMBRE','SEGURO','FEB','MAR','ABR','MAY','JUN','JUL','AGO','SEPT','OCT','DEUDA']
```

```python
def mayus(df):
    for col in df.select_dtypes(include="object").columns:
        if col in ["APELLIDO", "NOMBRE"]:
            df[col] = df[col].apply(str.title)
        # Aplicar capitalize a cada valor de la columna si es una cadena de texto
        else:
            df[col] = df[col].apply(str.capitalize)
    return df
```
