### TEMA

El Club de Handball Rio Negro es una organización deportiva amateur, fundada en 2023 y autogestionada por sus jugadores. Desde su creación, el club ha centrado sus esfuerzos en consolidarse y generar una estructura sostenible en categorías mixtas e individuales para jugadores de diversas edades y géneros. Sin embargo, debido a su corta trayectoria y el modelo de ingresos limitados a las cuotas de sus socios, enfrenta desafíos financieros importantes que ponen en riesgo su viabilidad futura.

Este proyecto tiene como objetivo llevar a cabo un análisis exploratorio de los datos financieros y de membresía del club. Para ello, se investigarán los balances de ingresos y egresos mensuales, así como la rotación de socios, la tasa de deserción, y el impacto financiero de los jugadores becados (aquellos exentos de la totalidad o parte de la cuota). A través de Python y sus librerías como Pandas, Matplotlib y Seaborn, el análisis permitirá evaluar la situación financiera actual del club y proyectar su viabilidad a mediano y largo plazo.
