### TÍTULO

Análisis Exploratorio de datos financieros y de membresía del Club de Handball Rio Negro

### TEMA

El Club de Handball Rio Negro es una organización deportiva amateur, fundada en 2023 y autogestionada por sus jugadores. Desde su creación, el club ha centrado sus esfuerzos en consolidarse y generar una estructura sostenible en categorías mixtas e individuales para jugadores de diversas edades y géneros. Sin embargo, debido a su corta trayectoria y el modelo de ingresos limitados a las cuotas de sus socios, enfrenta desafíos financieros importantes que ponen en riesgo su viabilidad futura.

Este proyecto tiene como objetivo llevar a cabo un análisis exploratorio de los datos financieros y de membresía del club. Para ello, se investigarán los balances de ingresos y egresos mensuales, así como la rotación de socios, la tasa de deserción, y el impacto financiero de los jugadores becados (aquellos exentos de la totalidad o parte de la cuota). A través de Python y sus librerías como Pandas, Matplotlib y Seaborn, el análisis permitirá evaluar la situación financiera actual del club y proyectar su viabilidad a mediano y largo plazo.

### HIPÓTESIS

La hipótesis central del proyecto se enfoca en la sostenibilidad económica del club y su capacidad para retener a los socios en el tiempo. Basándonos en los datos recopilados, se han definido las siguientes hipótesis principales y secundarias:

**1) Hipótesis principal:**

Sostenibilidad Financiera: Dado que el club depende casi exclusivamente de las cuotas de sus socios, la estructura actual de ingresos y egresos del Club de Handball Rio Negro no es suficiente para garantizar su sostenibilidad financiera a largo plazo. Esto implica que, sin una optimización de gastos o la incorporación de nuevas fuentes de ingresos, el club podría enfrentar dificultades económicas críticas en los próximos años.

**Hipótesis secundarias**

1) Impacto de los Jugadores Becados en los Balances del Club:

La presencia de un porcentaje significativo de jugadores becados afecta los ingresos regulares del club, disminuyendo su capacidad para cubrir los gastos operativos básicos. Se espera encontrar que, a mayor porcentaje de jugadores becados, menor será el balance neto de ingresos, lo que podría requerir un ajuste en la política de becas o en el esquema de cuotas.

2) Estacionalidad en la Rotación de Socios: 

Se asume que existe una estacionalidad en la incorporación y deserción de socios. Es decir, algunos meses del año (por ejemplo, después de períodos vacacionales) pueden mostrar mayores tasas de inscripciones y otros (fin de temporada o año) pueden reflejar un aumento en las bajas de socios. Identificar estos periodos es crucial para anticipar fluctuaciones de ingresos y planificar estratégicamente la captación y retención de miembros.

3) Análisis Detallado de la Retención y Perfil de los Jugadores

    a. Perfil Demográfico y de Participación de Socios Activos
        
    - Crear un perfil detallado de los jugadores, agrupando datos por categoría y género, junto con las métricas de retención y rotación. Esto incluirá:
        - Proporción de jugadores por género y categoría.
        - Edad promedio por categoría.
        - Duración promedio de membresía por categoría y género (evaluando la permanencia en el club).

    b. Identificación de Grupos con Mayor Retención y Rotación
        
    - Observar cuáles categorías y géneros presentan los mayores índices de retención y cuáles tienen la rotación más alta. Esto permitiría resaltar a los grupos con una gran base de miembros y una retención más sólida, mostrando a sponsors el compromiso de los jugadores en estos segmentos.

    c. Comportamiento de pago por segmento
    
    - Analizar el comportamiento de pago, identificando las categorías con mejor tasa de pagos completos y menores saldos pendientes. Esta información puede indicar el nivel de compromiso financiero de cada grupo, lo cual es atractivo para sponsors que buscan apoyar a segmentos con estabilidad y compromiso.

    d. Impacto de las Becas en la Retención
    
    - Evaluar cómo la concesión de becas afecta la retención de jugadores en las distintas categorías. Mostrar el impacto positivo de estas ayudas podría interesar a patrocinadores en apoyar un programa de becas, promoviendo la inclusión en el club.

    e. Análisis Temporal y de Estacionalidad

    - Analizar cómo varía la inscripción y deserción de jugadores a lo largo del año, y si ciertos meses presentan picos de altas o bajas. Esta información ayuda a demostrar patrones de participación activa y podría ser usada para hacer coincidir estrategias de patrocinio en períodos de alta visibilidad o interés en el club.

*Utilización de los Insights para Estrategia de Sponsors*: Basado en este análisis, el perfil detallado de los socios y su comportamiento de retención podrían posicionarse como argumentos sólidos al presentar una propuesta de patrocinio:

*Apoyo a la Comunidad y la Inclusión*: Mostrar a sponsors cómo su contribución podría facilitar que el club continúe siendo accesible a una variedad de jugadores en diferentes situaciones económicas (especialmente relevante para empresas interesadas en Responsabilidad Social Corporativa - RSC).

*Potencial de Visibilidad*: Segmentos de mayor retención y con menor rotación indican audiencias comprometidas, que garantizan una visibilidad continua para los sponsors.

*Datos sobre Comportamiento y Engagement*: Al conocer los segmentos con mejor cumplimiento de pagos y retención, los sponsors podrían identificar un perfil de audiencia responsable y comprometida, una ventaja al buscar patrocinar una comunidad deportiva sólida.

Con este enfoque, el análisis no solo ayuda a evaluar la sostenibilidad del club, sino que también permite construir una narrativa sólida para atraer posibles patrocinadores, mostrándoles el impacto y la visibilidad que sus aportes podrían tener en el club y en la comunidad.

4) Disparidad en la Retención por Categoría y Género: 

Se plantea la hipótesis de que en las categorías recreativas de niños (mini e infantil), sin objetivos de competencia, existe una alta afluencia y rotación de jugadores, debido a que estas categorías tienen un enfoque recreativo y no competitivo. Sin embargo, a medida que se avanza en las categorías competitivas (menores, cadetes, juveniles y mayores), se espera que aumente el compromiso de los jugadores, reduciendo proporcionalmente la rotación. Este análisis evaluará si, efectivamente, los jugadores en categorías competitivas muestran una mayor permanencia y compromiso, lo cual resultaría en una retención más alta en estos grupos en comparación con las categorías recreativas.

5) Proporción de Gastos y Necesidades de Optimización: 

En la estructura de egresos del club, los pagos a entrenadores y los seguros pueden representar un alto porcentaje de los gastos totales. Es probable que haya ciertos egresos que, si se optimizan o ajustan, podrían mejorar el balance financiero del club. La hipótesis sugiere que una redistribución de gastos y la posible reducción de costos en categorías como “materiales” o “extras” podrían aliviar el déficit, mejorando la salud financiera sin afectar el funcionamiento del club.
